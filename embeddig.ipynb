```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentence
sent=['the glass of milk',
      'the glass off juice',
      'the cup of tea',
      'I am good boy',
      'I am a good developer',
      'understanding the meaning of words',
      'your videos are good',]
```

```python
sent
```

```python
# define vocbulary size
voc_size = 10000
```

```python
## One_hot_representation
onehot_rep = [one_hot(words,voc_size) for words in sent]
onehot_rep
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_length=8
emb_doc = pad_sequences(onehot_rep,padding='pre',maxlen=sent_length)
print(emb_doc)
```

```python
voc_size = 10000  # Example vocabulary size (should be >0)
dim = 10        # Example embedding dimension (should be >0)
sent_length = 8 # Example sentence length

model = Sequential()
model.add(Embedding(voc_size, dim, input_length=sent_length))
model.compile(optimizer='adam', loss='mse')
```
